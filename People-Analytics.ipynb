```python
import pandas as pd
```

```python
df = pd.read_csv('HR Metrics - Sheet1.csv')
```

```python
df.head()
```

```python
df.columns
```

```python
# data cleaning
df.columns = df.columns.str.strip().str.lower().str.replace('_', '') # standardize column names
df = df.drop_duplicates() # remove duplicates
```

```python
print(df.columns.tolist())
```

```python
df.rename(columns={'employeeid': 'employee_id'}, inplace=True)
df.rename(columns={'jobrole': 'job_role'}, inplace=True)
df.rename(columns={'maritalstatus': 'marital_status'}, inplace=True)
df.rename(columns={'hiredate': 'hire_date'}, inplace=True)
df.rename(columns={'yearsatcompany': 'years_at_company'}, inplace=True)
df.rename(columns={'joblevel': 'job_level'}, inplace=True)
df.rename(columns={'monthlyincome': 'monthly_income'}, inplace=True)
df.rename(columns={'salaryhike%': 'salary_hike%'}, inplace=True)
df.rename(columns={'performancerating': 'performance_rating'}, inplace=True)
df.rename(columns={'trainingtimeslastyear': 'training_times_last_year'}, inplace=True)
df.rename(columns={'worklifebalance': 'work_life_balance'}, inplace=True)
df.rename(columns={'jobsatisfaction': 'job_satisfaction'}, inplace=True)
df.rename(columns={'environmentsatisfaction': 'environment_satisfaction'}, inplace=True)
df.rename(columns={'promotionlast5years': 'promotion_last_5years'}, inplace=True)
```

```python
df.info()
```

```python
df['hire_date'] = pd.to_datetime(df['hire_date'], dayfirst=True, format='mixed')
```

```python
df.isnull().sum()
```

```python
df = df.drop_duplicates()
```
